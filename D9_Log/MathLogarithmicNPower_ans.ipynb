此作業之目的是增加同學對自然對數Logarithmic 和指數 power 函數的認識

1. 以exp(a) 回傳 e 的 3次冪.

2. 以log(a, b)回傳以 10為底,100 的對數

```python
# 以 Python 程式碼展示
# exp() 和 log() 函數

# 載入 "math" 數學程式庫
import math

# 回傳 e 的 3次冪
print ("The e**3 value is : ", end="")
print (math.exp(3))

# 回傳 10,100 的對數
print ("The value of log 100 with base 10 is : ", end="")
print (math.log(100,10))
```

3. 以log2(a) 回傳以2為底的64的對數. 

4. 以log10(a) 回傳以10為底的1000的對數.

```python
# 以 Python 程式碼展示
# log2() 和 log10()

# 載入 "math" 數學程式庫
import math

# returning the log2 of 64
print ("The value of log2 of 64 is : ", end="")
print (math.log2(64))

# returning the log10 of 1000
print ("The value of log10 of 1000 is : ", end="")
print (math.log10(1000))
```

5. 以pow(a, b) 計算 4 的 3 次方 (4**3).

6. 以sqrt()  回傳49的平方根值.

```python
# 以 Python 程式碼展示
# pow() 和 sqrt()

# 載入 "math" 數學程式庫
import math

# returning the value of 4**3
print ("The value of 4 to the power 3 is : ", end="")
print (math.pow(4,3))

# returning the square root of 49
print ("The value of square root of 49 : ", end="")
print (math.sqrt(49))
```
